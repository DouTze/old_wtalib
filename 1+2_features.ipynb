







# **3. Features vs Previous Candlestick**  
### ***Increasing*** / ***MoveUp***  
**Look-back Len:**  
> 1  

**Rules:**  
> ***Close***[*idx] > ***Close***[*idx*-1]  

### ***Declining*** / ***MoveDown***  
**Look-back Len:**  
> 1  

**Rules:**  
> ***Close***[*idx] < ***Close***[*idx*-1]  

### ***RealBody GappingUp***  
**Look-back Len:**  
> 1  

**Rules:**  
> *min*(***Open***[*idx*], ***Close***[*idx*]) > *max*(***Open***[*idx*-1], ***Close***[*idx*-1])  

### ***RealBody GappingDown***  
**Look-back Len:**  
> 1  

**Rules:**  
> *max*(***Open***[*idx*], ***Close***[*idx*]) < *min*(***Open***[*idx*-1], ***Close***[*idx*-1])  

### ***With UpsideGap***  
**Look-back Len:**  
> 1  

**Rules:**  
> ***Low***[*idx*] > ***High***[*idx*-1]  

### ***With DownsideGap***  
**Look-back Len:**  
> 1  

**Rules:**  
> ***High***[*idx*] < ***Low***[*idx*-1]  

 





## **4.3** ***Within***  
**Notations**  
- *target*: ***Open***, ***High***, ***Low***, ***Close***, ***RealBody***, ***Amplitude***, ***LowerShadow***, ***UpperShadow***, ***Shadows***, ***BodyRange***, ***HighLowRange***  
- *reference*: ***BodyRange***, ***HighLowRange***  
- *Paras* := [*average_period*:***float***, *average_factor*:***float***]  

### ***Within***  
**Parameters:**  
> ***None***

**Look-back Len:**  
> 0   

**pre-definitions:**  
- *t_idx* := *idx* of *target*  
- *r_idx* := *idx* of *reference*  
- *lower-bound* = *reference*[*r_idx*].*lower-bound*  
- *upper-bound* = *reference*[*r_idx*].*upper-bound*  

**Rules:**  
> *target* is **Range-Type** ? *target*.*lower-bound* >= *lower-bound* and *target*.*upper-bound* <= *upper-bound* : *target* >= *lower-bound* and *target* <= *upper-bound*  

### ***Within/Near***  
**Parameters:**  
> *Paras* := *Paras*[***Near***]  

**Look-back Len:**  
> *Paras*.*average_period*  

**pre-definitions:**  
- *t_idx* := *idx* of *target*  
- *r_idx* := *idx* of *reference*  
- *threshold* = ***Amplitude***[*r_idx*-*Paras*.*average_period*: *r_idx*].*mean*() * *Paras*.*factor*  
- *lower-bound* = *reference*[*r_idx*].*lower-bound* - *threshold*  
- *upper-bound* = *reference*[*r_idx*].*upper-bound* + *threshold*  

**Rules:**  
> *target* is **Range-Type** ? *target*.*lower-bound* >= *lower-bound* and *target*.*upper-bound* <= *upper-bound* : *target* >= *lower-bound* and *target* <= *upper-bound*
